```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re # Regular Expression
import string # contains all string value
from sklearn.model_selection import train_test_split # Use the train the dataset
from sklearn.metrics import accuracy_score # Check the accuracy values
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression # Binary values 0 and 1
from sklearn.feature_extraction.text import TfidfVectorizer # converts num to text tfidf - Term Frequency -Inverse Documents Frequency
from sklearn.ensemble import RandomForestClassifier # Help to identify the different tree (randomforest is a type of decisiontree but class.. model)
from sklearn.naive_bayes import MultinomialNB # text classification
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'https?://\S+', '', text)
    text = re.sub(r'<.*?>', '', text)
    text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub(r'\n', ' ', text)
    text = re.sub(r'\w*\d\w*', '', text)
    return text
```

```python
vectorizer=TfidfVectorizer(stop_words='english',max_df=0.7)
```

```python
x_test=vectorizer.transform(x_test)
```

```python
model=LogisticRegression(max_iter=1000)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```
